## Time Differences

```python
# Sort by user and then time
sorted_ = data1.sort_values(['user', 'timestamp'])
sorted_['timestamp'] = pd.to_datetime(sorted_['timestamp'])
# Get the difference for each user for eached placed pixel
sorted_['time_diff'] = sorted_.groupby('user')['timestamp'].diff()
```

```python
def time_convert(x):
    x = str(x).replace('days ', '')
    (days, time) = x.split(' ')
    (h,m,s) = time.split(':')
    results = float(days) * 86400 + float(h) * 3600 + float(m) * 60 + float(s)
    return results
```

```python
sorted_c = sorted_.copy()

sorted_c = sorted_.dropna()
# Convert time difference to seconds
sorted_c['time_diff'] = sorted_c['time_diff'].apply(lambda x: time_convert(x))

# Ignore pauses
sorted_c = sorted_c[sorted_c['time_diff'] < 28800]

# Take the mean
mean = sorted_c.groupby('user')['time_diff'].mean()
mean = mean.to_frame().reset_index().dropna()
mean.rename(columns = {'time_diff':'mean_time_diff'}, inplace = True)

# Take the variance for each user
variance = sorted_c.groupby('user')['time_diff'].var()
variance = variance.to_frame().reset_index().dropna()
variance.rename(columns = {'time_diff':'var_of_time_diff'}, inplace = True)

variance = variance.merge(mean, on = 'user')
```

```python
variance_c = variance.copy()

# Merge variance and user count tables
variance_c = variance.merge(user_counts, on = 'user')
# Filter for low variance, high visits
low_variance = variance_c[(variance_c['var_of_time_diff'] < 10) & (variance_c['visits'] > 12)]
# List of low variance users
low_variance_list = low_variance['user'].to_list()
```
